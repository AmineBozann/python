```python
# ğŸ“˜ METÄ°N ANALÄ°Z ARACI
# Bu program bir metin dosyasÄ±nÄ± okur, iÃ§eriÄŸini analiz eder
# ve kelime sÄ±klÄ±klarÄ±, cÃ¼mle sayÄ±sÄ± gibi istatistikleri gÃ¶sterir.

import string                # Noktalama iÅŸaretlerini temizlemek iÃ§in
from collections import Counter   # Kelime sÄ±klÄ±klarÄ±nÄ± hesaplamak iÃ§in
import matplotlib.pyplot as plt   # Grafik Ã§izmek iÃ§in

# ğŸ”¹ 1. AdÄ±m: DosyayÄ± okuyan bir fonksiyon
def metin_oku(dosya_adi):
    """
    Belirtilen dosyayÄ± okur ve iÃ§eriÄŸini bir metin (string) olarak dÃ¶ndÃ¼rÃ¼r.
    """
    with open(dosya_adi, "r", encoding="utf-8") as dosya:
        return dosya.read()

# ğŸ”¹ 2. AdÄ±m: Metni temizleyen bir fonksiyon
def metin_temizle(metin):
    """
    Noktalama iÅŸaretlerini kaldÄ±rÄ±r, tÃ¼m harfleri kÃ¼Ã§Ã¼k yapar.
    """
    temiz_metin = metin.lower()
    for nokta in string.punctuation:
        temiz_metin = temiz_metin.replace(nokta, "")
    return temiz_metin

# ğŸ”¹ 3. AdÄ±m: Kelime analiz fonksiyonu
def kelime_analiz_et(metin):
    """
    Metindeki kelimeleri sayar, sÄ±klÄ±klarÄ±nÄ± dÃ¶ndÃ¼rÃ¼r.
    """
    kelimeler = metin.split()
    kelime_sayisi = len(kelimeler)
    ort_uzunluk = sum(len(k) for k in kelimeler) / kelime_sayisi
    kelime_sikliklari = Counter(kelimeler)
    return kelimeler, kelime_sayisi, ort_uzunluk, kelime_sikliklari

# ğŸ”¹ 4. AdÄ±m: CÃ¼mle analizi
def cumle_analiz_et(metin):
    """
    Noktaya gÃ¶re cÃ¼mleleri ayÄ±rÄ±r ve toplam cÃ¼mle sayÄ±sÄ±nÄ± dÃ¶ndÃ¼rÃ¼r.
    """
    cumleler = [c for c in metin.split('.') if c.strip() != ""]
    return len(cumleler)

# ğŸ”¹ 5. AdÄ±m: SonuÃ§larÄ± yazdÄ±ran fonksiyon
def sonuc_goster(kelime_sayisi, cumle_sayisi, ort_uzunluk, kelime_sikliklari):
    """
    Analiz sonuÃ§larÄ±nÄ± ekrana yazdÄ±rÄ±r.
    """
    print("\nğŸ” METÄ°N ANALÄ°ZÄ° SONUÃ‡LARI ğŸ”")
    print(f"Toplam kelime sayÄ±sÄ±   : {kelime_sayisi}")
    print(f"Toplam cÃ¼mle sayÄ±sÄ±    : {cumle_sayisi}")
    print(f"Ortalama kelime uzunluÄŸu: {ort_uzunluk:.2f} karakter")

    print("\nğŸ“ˆ En sÄ±k geÃ§en 10 kelime:")
    for kelime, adet in kelime_sikliklari.most_common(10):
        print(f"{kelime:<15} {adet} kez")

# ğŸ”¹ 6. AdÄ±m: Kelime sÄ±klÄ±klarÄ±nÄ± grafikle gÃ¶steren fonksiyon
def grafik_ciz(kelime_sikliklari):
    """
    En sÄ±k geÃ§en 10 kelimeyi Ã§ubuk grafik olarak gÃ¶sterir.
    """
    en_cok = kelime_sikliklari.most_common(10)
    kelimeler = [k for k, _ in en_cok]
    adetler = [a for _, a in en_cok]

    plt.figure(figsize=(8,5))
    plt.bar(kelimeler, adetler)
    plt.title("En SÄ±k GeÃ§en 10 Kelime")
    plt.xlabel("Kelimeler")
    plt.ylabel("GeÃ§iÅŸ SayÄ±sÄ±")
    plt.show()

# ğŸ”¹ 7. AdÄ±m: Ana algoritma akÄ±ÅŸÄ±
def ana_program():
    """
    TÃ¼m iÅŸlemleri sÄ±rasÄ±yla yapan ana fonksiyon.
    """
    dosya_adi = input("Analiz edilecek metin dosyasÄ±nÄ±n adÄ±nÄ± girin (Ã¶rnek: metin.txt): ")

    try:
        metin = metin_oku(dosya_adi)
        temiz = metin_temizle(metin)
        kelimeler, kelime_sayisi, ort_uzunluk, kelime_sikliklari = kelime_analiz_et(temiz)
        cumle_sayisi = cumle_analiz_et(metin)

        sonuc_goster(kelime_sayisi, cumle_sayisi, ort_uzunluk, kelime_sikliklari)
        grafik_ciz(kelime_sikliklari)
    except FileNotFoundError:
        print("âŒ Dosya bulunamadÄ±. LÃ¼tfen dosya adÄ±nÄ± doÄŸru girin.")

# ğŸ”¹ 8. AdÄ±m: ProgramÄ±n Ã§alÄ±ÅŸtÄ±rÄ±lmasÄ±
if __name__ == "__main__":
    ana_program()
```
